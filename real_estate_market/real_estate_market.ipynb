В распоряжении данные сервиса Яндекс.Недвижимость — архив объявлений о продаже квартир в Санкт-Петербурге и соседних населённых пунктах за несколько лет. Нужно научиться определять рыночную стоимость объектов недвижимости. Задача — установить параметры. Это позволит построить автоматизированную систему: она отследит аномалии и мошенническую деятельность. 

По каждой квартире на продажу доступны два вида данных. Первые вписаны пользователем, вторые получены автоматически на основе картографических данных. Например, расстояние до центра, аэропорта, ближайшего парка и водоёма.

## Предобработка данных

### Переименования столбцов

Для переименования колонки `cityCenters_nearest` воспользуемся методом *rename()*, передав в качестве аргументов:
- словарь из старого и нового значения; 
- *axis = 1*, чтобы переименование произошло в колонках;
- *inplace = True*, чтобы произошла перезапись таблицы.

```python
real_estate_data.rename({'cityCenters_nearest': 'city_centers_nearest'}, axis=1, inplace=True) #меняем название колонки
real_estate_data.columns #проверяем изменения в названии колонок
```

**Итог**: Теперь все названия соответствуют правилам хорошего стиля.


### Обработка пропусков

В таблице присутствуют данные двух видов: введенные пользователями и полученные автоматически на основе картографических данных. Обработку пропусков также произведем в два блока.

#### Обработка пропусков, внесенных пользователями

**Причиной** пропусков в этом виде данных является тот факт, что пользователь не внес значения по необходимому параметру. К этому виду относятся колонки с пропусками `ceiling_height`, `floors_total`, `living_area`, `is_apartment`, `kitchen_area`, `balcony`, `days_exposition`.
   * В колонках `is_apartment` и `balcony` пропущенные значения свидетельствуют о том, что в квартире отсутствуют балконы или что квартира не относится к типу *апартаменты*. Соответственно, в обоих колонках можно заменить пропущенные значения на ноль.
   * Процент пропущенных значений в колонке `floors_total` равен 0,36%, следовательно, эти данные можно удалить, не боясь исказить дальнейшие исследования.
   * Пропуски в колонке `days_exposition` свидетельствуют о том, что квартира ещё не снята с продажи, объявление ещё опубликовано. Поскольку замена на какое-либо характерное значение может повлечь ошибку в дальнейших расчетах, оставим пока что пропуски не обработанными в данной колонке.
   * В колонке `living_area` находятся данные, которые нужны для дальнейших расчетов и исследования, а процент пропущенных значений достаточно высокий. Поэтому данные в колонке необходимо восстановить характерным значением - средним арифметическим или медианой по сгруппированным данным другой ячейки. Группировать данные будем по колонке `rooms`, поскольку количество комнат должно влиять на жилую площадь. Аналогичным образом заменим пропущенные значения в колонке `ceiling_height` и `kitchen_area`.
   
Заменим пропущенные значения в колонках `is_apartment` и `balcony`, применив метод *fillna()* и передав ему в качестве аргумента словарь с ключ-значением - *столбец-значение для замены*.

```python
real_estate_data.fillna({'is_apartment': 0, 'balcony': 0}, inplace=True) # заменяем пропуски и перезаписываем таблицу
real_estate_data.isna().sum() #проверяем, что пропущенные значения успешно заменены.
```

Количество пропусков в колонках `is_apartment` и `balcony` равно нулю, значит замена прошла успешно.

Удалим строки, в которых есть пропущенное значение в колонке `floors_total`. Для этого воспользуемся методом *dropna()* с аргументами *subset* равным названию колонки, по которой ищутся пропуски и *inplace* = True для перезаписи таблицы.

```python
real_estate_data.dropna(subset = ['floors_total'], inplace=True) #удаляем строки с пустыми значениями в колонке floors_total.
real_estate_data.isna().sum() #проверяем, что пропущенных значений в колонке не осталось
```

Пропуски в колонке `floors_total` успешно удалены из таблицы.

Пропущенное значение в колонке `days_exposition` говорит о том, что квартира ещё находится в продаже. В дальнейшем этот столбец пригодится для определения скорости продажи квартиры. Замена пропусков на характерные значения в дальнейшем может запутать и привести к ошибке при расчетах. Поэтому оставим пропуски в этой колонке пока необработанными.

Для обработки пропусков в колонке `living_area` проверим, зависит ли жилая и общая площадь от количества комнат в квартире. Для этого создадим сводную таблицу, группированную по значениям столбца `rooms` и найдем медиану значений в колонках `total_area` и `living_area`. Выбор медианы связан с тем, что она меньше подвержана влиянию вбросов со стороны данных (очень больших или очень маленьких значений). Создавать сводную таблицу будем при помощи метода *pivot_table*, в аргументы которого передадим соответствующие столбцы и значения.

```python
real_estate_data.pivot_table(index='rooms', values = ['total_area', 'living_area'], aggfunc = 'median') #создание сводной таблицы
```

Исходя из полученной сводной таблицы, можно сделать вывод, что при значениях от 0 до 9 наблюдается прямая зависимость между величинами: с увеличением числа комнат в квартире, растет как общая площадь, так и жилая. При значениях `rooms` выше 9 начинается разброс данных. Это может быть связано с тем, что количество наблюдений для квартир с большим числом квартир недостаточно. Проверим это, вызвав метод *value_counts()* с поиском и подсчетом уникальных значений в колонке `rooms`.

```python
real_estate_data['rooms'].value_counts()
```

Все верно, для квартир с числом комнат от 10 и выше количество наблюдений не превышает 3. Поэтому в значениях начинает появляться разброс.

Исходя из прямой зависимости между числом комнат и жилой площадью, можно заменить пропущенные значения на медиану по сгруппированным значениям числа комнат в квартире.
Создадим функцию *replacement_passes*, которая в качестве аргументов принимает таблицу и столбец с пропущенными значениями, а в теле функции при помощи цикла меняет пропущенные значения в колонке на значение медианы.

```python
def replacement_passes (data, column): #создаем функцию
    # в теле функции создаем переменную, в которой списком сохраняем медианные значения по колонке, сгруппированные по колонке rooms
    # создаем переменную, в которой списком сохраняем уникальные значения колонки rooms для пропущенные значений в колонке
    # циклом меняем пропущенные значения на медиану по количеству комнат соответственно.
    median_group = data.groupby('rooms')[column].median()
    rooms_number = data[data[column].isna()]['rooms'].unique()
    for room in rooms_number:
        data.loc[data['rooms'] == room, column] = data.loc[data['rooms'] == room, column].fillna(median_group[room])
```

```python
replacement_passes(real_estate_data, 'living_area') #вызываем функцию для замены пропусков в колонке living_area
real_estate_data.isna().sum() #проверяем, что пропущенных значений в колонке не осталось
```

Пропущенные значения в колонке `living_area` обработаны. Повторим вызов функции для колонок `ceiling_height` и `kitchen_area`.

```python
replacement_passes(real_estate_data, 'kitchen_area') #вызов функции для колонки kitchen_area
replacement_passes(real_estate_data, 'ceiling_height') #вызов функции для колонки ceiling_height

real_estate_data.isna().sum() #проверяем, что пропущенные значения обработались
```

Количество пропущенных значений в колонках `kitchen_area` и `ceiling_height` уменьшилось, но не обработались полностью. Посмотрим, для какого количества комнат пропуски в колонках не заменились на медиану. ### Категоризация квартир по этажу
Произведем категоризацию данных для этажа квартиры. Если квартира находится на первом этаже, присвоим ей категорию `первый`, если на последнем этаже - `последний`. Если квартира не находится ни на первом, ни на последнем этаже, присвоим категорию `другой`. Создадим функцию категоризации, которая на входе будет принимать строку таблицы, а на выходе возвращать категорию квартиры по этажу.

```python
def floor_group(row):
    """
    Возвращает группу этажа по значению этажа floor, используя правила:
    - 'первый', если floor == 1 этаж;
    - 'последний', если floor == последний этаж в здании (== real_estate_data['floors_total']);
    - 'другой' — во всех остальных случаях.
    """
    if row['floor'] == 1:
        return 'первый'
    if row['floor'] == row['floors_total']:
        return 'последний'
    return 'другой'
```

Вызовем функцию для таблицы при помощи метода *apply()*. В аргументе метода укажем, что принимать необходимо строки (*axis=1*)

```python
real_estate_data['floor_category'] = real_estate_data.apply(floor_group, axis=1)
```

#### Этаж, день недели, месяц, год

Поищем взаимосвязь между ценой квартиры в колонке `last_price` и колонками `weekday_exposition`, `month_exposition`, `year_exposition`, `floor_category`. Для этого напишем функцию, которая будет строить сводную таблицу, группировав значения заданной колонки и находя медиану значений последней цены.

```python
def price_study(column, data=filter_estate_data): #создаем функцию. Задаем параметр по умолчанию data = filter_estate_data
    (data
     .pivot_table(index = column, values = 'last_price', aggfunc = 'median')
     .plot(style='o-', grid = True, figsize = (12,5)))
```

в цикле запускаем функцию для списка с нужными колонками.

```python
for column in ['weekday_exposition', 'month_exposition', 'year_exposition', 'floor_category', 'rooms']:
    price_study(column)
```

**Выводы:**

#### Анализ квартир в центре

Для начала из таблицы *spb_estate_data* выделим сегмент квартир, которые находятся в центр города. Для этого методом *query()* создадим условие, в котором расстояние в километрах меньше/равно 6. Данные сохраним в переменной *spb_centers_estate_data*.

```python
spb_centers_estate_data = spb_estate_data.query('city_centers_nearest_km <= 6')
```

Создадим функцию *centers_apartments*, которая будет принимать на входе название колонки и возвращать две гистограммы: первая - для квартир в центре, вторая - для всего города. Таким образом, произведем сравнение между квартирами в центре и во всем городе.

```python
def centers_apartments(column, data_one=spb_centers_estate_data, data_two=spb_estate_data): #создание функции
    data_one.plot(
    kind='hist',
    y=column,
    histtype='step',
    linewidth=5,
    alpha=0.7,
    label=column + ' center',
    grid=True,
    legend=True
    )
    data_two.plot(
    kind='hist',
    y=column,
    histtype='step',
    linewidth=5,
    alpha=0.7,
    label=column + ' all_city',
    grid=True,
    legend=True
    ) 
```

В цикле вызовем функцию для колонок `rooms`, `total_area`, `ceiling_height` и `last_price`.

```python
for column in ['rooms', 'total_area', 'ceiling_height', 'last_price']:
    centers_apartments(column)
```

**Выводы:**
- Для параметров число комнат, общая площадь и цена продажи пик на гистограме сместился вправо. Следовательно, в центре находятся квартиры с большим числом комнат, большие по площади (помним о корреляции между числом комнат и площадью) и более дорогие (что также подтверждается корреляцией между величина цена продажи-общая площадь, цена продажи-расстояние до центра).
- Для гистограммы по параметру высота потолков один из пиков совпадает с пиком для всего города. Однако, для квартир в центре также имеются пики на значениях правее пика гистограммы всего города. Например, виднеется пик на значении 4 метра. Это говорит о том, что квартиры в центре выше по высоте потолка.

Исследуем факторы (этаж, число комнат, удаленность от центра, дата размещения), которые влияют на цену квартиры.
