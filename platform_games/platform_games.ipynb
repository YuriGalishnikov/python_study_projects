# Описание проекта

Интернет-магазин «Стримчик» продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). В проекте необходимо  выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и необходимо спланировать кампанию на 2017-й.

В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков». Существуют следующие возрастные категории:

- **«EC» («Early childhood») — «Для детей младшего возраста»:** Игра подходит для детей от 3 лет и не содержит материалов, которые родители могли бы счесть неподходящими. Продукты, получившие данный рейтинг, изначально разрабатываются для детей и обычно представляют собой развивающие игры.
- **«E» («Everyone») — «Для всех»:** Содержание вполне подходит для возрастной категории от 6 лет. Такие игры могут понравиться и взрослым. Игры с этим рейтингом могут содержать минимальное насилие, в основном «мультяшного» характера.
- **«E10+» («Everyone 10 and older») — «Для всех от 10 лет и старше»:** Проекты с данным рейтингом могут содержать или немного больше мультипликационного или мягкого насилия, или несколько откровенные сцены, или минимальное количество крови.
- **«T» («Teen») — «Подросткам»: Игра подходит для лиц от 13 лет**. Проекты из данной категории могут содержать насилие, непристойные сцены, грубый юмор, в меру откровенное сексуальное содержимое, кровь или нечастое использование ненормативной лексики.
- **«M» («Mature») — «Для взрослых»:** Материалы игры не подходят для подростков младше 17 лет. Проекты с данным рейтингом могут содержать достаточно жестокое насилие, большое количество крови с расчленением, непристойные сексуальные сцены или грубую ненормативную лексику, нежелательную для младшей аудитории.
- **«AO» («Adults Only 18+») — «Только для взрослых»:** Содержание игры только для взрослых старше 18 лет. Продукты из данной категории могут содержать длительные сцены жестокого насилия и/или очень откровенное сексуальное содержимое, а также сцены с обнажением.
- **«RP» («Rating Pending») — «Рейтинг ожидается»:** Продукт был отправлен в ESRB и ожидает присвоения рейтинга. Данный логотип используется только на рекламных презентациях и в демо-версиях игр до официальной даты выпуска в продажу.

**Примечание:**
Данные за 2016 год могут быть неполными.

**Цель исследования** - проверить две гипотезы:

1. Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

## Подготовка данных

### Обработка названий колонок

Для обработки названий колонок (приведения к нижнему регистру) воспользуемся методом библиотеки **Pandas** *str.lower()* и перезапишем названия колонок.

```python
games_data.columns = games_data.columns.str.lower() # приведение к нижнему регистру
games_data.columns # выводим список названий для проверки
```

Все названия приведены к нижнему регистру и соответствуют правилам хорошего стиля.

### Обработка типов данных

При помощи метода *astype()* приведим тип данных в колонках `year_of_release` и `critic_score` к целочисленному типу. В качестве аргумента передадим методу тип *Int64*, чтобы он смог обработать данные при наличие пропусков в колонке.

```python
# преобразование типа данных и перезапись данных в колонке
games_data[['year_of_release', 'critic_score']] = games_data[['year_of_release', 'critic_score']].astype('Int64') 
```

При помощи метода *to_numeric()* приведем данные в колонке `user_score` к числовому типу, а именно float. Как уже было сказано выше, в колонке `user_score` встречается значение `tbd`. В английском языке это сокращение от "To Be Determined (Decided)", что в переводе означает "будет определено (решено). В данном случае это значение в колонке `user_score` означает, что пользовательская оценка по данной игре еще не посчитана. Соответственно, мы оценки не знаем, значит значения можно заменить на NaN. При привидении к числовому типу при помощи метода *to_numeric()* воспользуемся аргументом метода *errors* и передадим ему значение *coerce*. В этом случае, все значения, которые метод не сможет привести к числу, он заменит на NaN, а также обработает без ошибки уже имеющиеся пропуски в этой колонке.

```python
# преобразование типа данных и перезапись данных в колонке
games_data['user_score'] = pd.to_numeric(games_data['user_score'], errors='coerce')
```

```python
games_data.info() # проверяем замену типов данных в колонках
```

Во всех колонках тип данных соответствует приведенным значениям.

### Обработка пропусков

Проверим, какую информацию несут строки с пропущенными значениями в колонке с названиями игр. Для этого отфильтруем таблицу по пропущенным значения в колонке `name`.

```python
games_data[games_data['name'].isna()] # выводим строки с пропущенными значениями в колонке name
```

Данные строки имеют информацию только о платформе, дате релиза и продажах. Стоит отметить, что количество пропусков в колонке `name` и в колонке `genre` совпадает - по два пропуска, а также эти пропуски находятся в одних и тех же строках. Процент пропусков составляет чуть более 0,01%. Данные строки можно удалить без потери качества данных (просто запомним, что в 1993 году было две игры на платформу GEN). Воспользуемся методом *dropna()*. В качестве аргумента *subset* передадим колонку `name`, а аргумента *inplace()* значение True для перезаписи данных.

```python
games_data.dropna(subset=['name'], inplace=True) # удаляем строки с пропусками в колонке name 
```

```python
games_data.info() # проверяем обновленную информацию по таблице
```

Пропусков в колонках `name` и `genre` не осталось, также уменьшилось количество строк в таблице на 2.

Пропуски в колонках `critic_score`, `user_score` не имеет адекватной возможности быть обработаны. Так, например, отсутствуют зависимости между пропуском в колонке и годом релиза игры или платформой, на которую выпускалась игра. Присваивать среднее или медианное значение по определенной группе (например, жанру) не выглядит оправданным, поскольку это может отразиться на действительности результатов. Также отсутствует возможность восстановить утерянные данные. Количество пропусков имеет слишком высокий процент, поэтому при удалении строк с пропущенными значениями, будет утеряно большое количество полезной информации и последующие расчеты и выводы могут быть некорректны. Следовательно, удалить данные мы тоже не можем. Оставим пропущенные значения в данных колонках как есть.

В колонке `rating` пропуски могут свидетельствовать об отсутствии рейтинга: так многие игры выпускаются вообще без данного рейтинга. Заменим пропущенные значения в данной колонке на *unknown*.

Причиной пропусков в колонках `critic_score` и `user_score` может быть отсутствие данных по оценкам для каких-то конкретных игр: возможно, оценки для этих игр не собирались или были утеряны в процессе сбора/занесения данных в таблицу.

Пропущенные значения в колонке `year_of_release` есть возможность восстановить: для каждой игры найти информацию в интернете о дате релиза и занести в таблицу. Однако, это достаточно затратный по времени процесс, а количество пропусков составляет чуть более 1,5%. Такое количество пропусков не скажется на корректности результатов, следовательно, их также оставим без обработки. Причины пропусков носят аналогичный характер.

Также нельзя исключать технические сбои при чтении, сборе и/или хранении таблицы, из-за которых были утеряны данные.

```python
games_data['rating'].fillna('unknown', inplace=True) # замена пропусков в колонке rating на unknown
```

Создадим функцию **top_platform_analysis**, которая будет рисовать график распределения суммарных продаж по годам:
- В качестве аргумента функция будет принимать название платформы (переменная *console*).
- В теле функции будет создаваться сводная таблица для этой консоли из таблицы `games_data`. В сводной таблице будут сохранены сумма по продажам за год и количество выпущенных игр для этой платформы за год. Далее будут рисоваться 2 столбчатые диаграммы на одном графике с распределение суммарных продаж и количеством выпущенных игр за год.

```python
def top_platform_analysis(console): # создание фукнции
    # создание сводной таблицы с информацией о количестве выпущенных за год игр и суммарными продажи за год для конкретной платформы
    new_data = (games_data.query('platform == @console')
                    .pivot_table(index='year_of_release', values=['name', 'all_sales'], aggfunc={'name':'count', 'all_sales':'sum'})
                    .reset_index())
    # создание столбчатых диаграмм с распределение показателей   
    fig, ax = plt.subplots() # создание области для графика

    ax.bar(new_data['year_of_release'], new_data['name']) # диаграмма для количества игр по годам
    ax.bar(new_data['year_of_release'], new_data['all_sales']) # диаграмма для суммарных продаж по годам

    ax.set_facecolor('seashell')
    fig.set_facecolor('floralwhite')
    fig.set_figwidth(12)    #  ширина Figure
    fig.set_figheight(6)    #  высота Figure
    ax.set_title(f"Количество и продажи игр на {console} по годам")

    plt.show() # вывод графика
    
    display(new_data) # вывод сводной таблицы
```

Теперь при помощи цикла вызовем функцию для 6ти отобранных популярных платформ.

```python
for platform in ["PS2", "X360", "PS3", "Wii", "DS", "PS"]: # создание цикла
    top_platform_analysis(platform) # вызов функции
```

На графиках оранжевым цветом обозначено количество продаж за год (в млн экземпляров), синим - количество выпущенных игр. 

Характерный срок существования платформы равен 10 годам. При этом период активного развития и/или стабильно высокого интереса к платформе составляет примерно 5-6 лет (для некоторых платформ меньше). После этого срока количество выпускаемых на платформу игр и, соответственно, продажи начинают падать.
Для платформы DS имеется выброс в виде 1 игры, выпущенной в 1985 году. После этого активное использование платформы DS началось только в 2004 году.

### Лидеры продаж в актуальном периоде

Создадим своднуб таблицу при помощи метода *pivot_table()* для таблицы за актуальный период, чтобы отобрать потенциально прибыльные платформы. В качестве аргументов передадим:
- *index* - список с колонками `platform` и `year_of_release`;
- *values* - список с колонками `all_sales` и `name`;
- *aggfunc* - словарь функций: для колонки `name` - `count`, для колонки `all_sales` - `sum`.

Аналогичным образом преобразуем индексы, чтобы они стали новыми колонками при помощи метода *reset_index()* без аргументов.

```python
# создание сводной таблицы
actual_games_data_pivot_table = (actual_games_data
                                 .pivot_table(index=['platform', 'year_of_release'], values=['name', 'all_sales'], aggfunc={'name':'count', 'all_sales':'sum'})
                                 .reset_index())
actual_games_data_pivot_table.head(10) # вывод таблицы
```

При помощи цикла, который будет проходить по уникальным значениям колонки `platform` полученной ранее сводной таблицы `actual_games_data_pivot_table` построим графики зависимости суммарных продаж за год и количества выпущенных за год игр для каждой платформы. 

```python
for i in actual_games_data_pivot_table['platform'].unique(): # создание цикла
    # сохраняем данные для одной платформы, отфильтровав при помощи метода query()
    platform_data = actual_games_data_pivot_table.query('platform == @i') 
    # строим графики
    fig, ax = plt.subplots() # создание области для графика

    ax.plot(platform_data['year_of_release'], platform_data['name'], label='Количество игр') # график зависимости количества игр от года
    ax.plot(platform_data['year_of_release'], platform_data['all_sales'], label='Продажи игр') # график зависимости суммарных продаж игр от года

    fig.set_figwidth(12) # ширина Figure
    fig.set_figheight(6) # высота Figure
    fig.set_facecolor('mintcream')
    ax.set_facecolor('whitesmoke')
    ax.set_title(f"Продажи за актуальный период на платформе {i}")
    ax.legend() # создание легенды

    plt.show() # вывод графика
```

На графиках оранжевым цветом обозначен показатель продаж игр в млн штук, синим - количество выпускаемых в год игр.

Потенциально прибыльные платформы на 2017 год: **3DS**, **PS4**, **PSV**, **WiiU**, **XOne**.

Для платформ **PS4** и **XOne** количество выпущенных игр растет от года к году. Падение продаж в 2016 году может являться следствием двух факторов: рост количество выпущенных игр в 2016 году по сравнению с 2015 несколько ниже или пользователи по каким-то третий факторам не смогли позволить себе купить больше игр в 2016 году. ## Проверка гипотез <a id='fifth'></a>

### Проверка гипотезы №1

Проверим гипотезу, что средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Возьмем две совокупности: первая - пользовательские рейтинги платформы Xbox One, вторая - пользовательские рейтинги платформы PC. Две эти совокупности являются независимыми, поскольку между ними отсутствуют связь, при которой можно было бы установить гоморфную пару.

Воспользуемся специальным методом для проверки гипотезы о равенстве среднего двух генеральных совокупностей: *scipy.stats.ttest_ind (array1, array2,equal_var)*. В качестве аргументов `array1` и `array2` будем передавать отфильтрованные по платформе данные колонки `user_score`. В качестве аргумента `equal_var` передадим False.

В качестве нулевой гипотезы (Н0) для статистического теста используем гипотезу о равенстве средних пользовательских рейтингов платформ Xbox One и PC. В качестве альтенативной (Н1) - гипотезу о неравенстве средних пользовательских рейтингов платформ.

```python
# статистический тест для проверки равенства средних пользовательских рейтингов платформ Xbox One и PC

alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    actual_games_data.query('platform == "XOne"')['user_score'], 
    actual_games_data.query('platform == "PC"')['user_score'], 
    equal_var=False, nan_policy='omit')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Отвергая гипотезу о равенстве среднего двух генеральных совокупностей, можно утвержать, что эти средние различаются. Гипотеза №1 не верна.

**Примечание:**

### Проверка гипотезы №2

Проверим гипотезу, что средние пользовательские рейтинги жанров Action и Sports разные.

Возьмем две совокупности: первая - пользовательские рейтинги жанра Action, вторая - пользовательские рейтинги жанра Sports. Две эти совокупности являются независимыми, поскольку между ними отсутствуют связь, при которой можно было бы установить гоморфную пару.

Проведем аналогичный статистический тест для проверки гипотезы о равенстве среднего двух генеральных совокупностей: *scipy.stats.ttest_ind (array1, array2,equal_var)*. В качестве аргументов `array1` и `array2` будем передавать отфильтрованные по жанру данные колонки `user_score`. В качестве аргумента `equal_var` передадим False.

В качестве нулевой гипотезы (Н0) для статистического теста используем гипотезу о равенстве средних пользовательских рейтингов жанров Action и Sports. В качестве альтенативной (Н1) - гипотезу о неравенстве средних пользовательских рейтингов жанров.

```python
# статистический тест для проверки равенства средних пользовательских рейтингов платформ Xbox One и PC

alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвегнем гипотезу

results = st.ttest_ind(
    actual_games_data.query('genre == "Action"')['user_score'], 
    actual_games_data.query('genre == "Sports"')['user_score'], 
    equal_var=False, nan_policy='omit')

print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Отвергая гипотезу о равенстве среднего двух генеральных совокупностей, можно утвержать, что эти средние различаются. Гипотеза №2 верна.
